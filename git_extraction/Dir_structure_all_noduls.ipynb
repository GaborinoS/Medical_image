```python
import pandas as pd
import numpy as np
import os
import sys
```

```python
dir_all_data = 'G:/Dokumente/MedicalIM/Lung_Cancer/manifest-1600709154662/LIDC-IDRI'
all_patients = os.listdir(dir_all_data)

folderpaths_patients = []

index =1
for index in range(len(all_patients)):
    x = os.listdir(dir_all_data + '/' + all_patients[index])

    res = []
    res2 = []

    for i in x:
        y = os.listdir(dir_all_data + '/' + all_patients[index] + '/' + i)
        res2.append(y)
        for j in y:
            z = os.listdir(dir_all_data + '/' + all_patients[index] + '/' + i + '/' + j)
            res.append(z)

            #print(len(z))

    ind = np.where(np.array([len(x) for x in res]) == max([len(x) for x in res]))[0][0]

    string = f'{dir_all_data}/{all_patients[index]}/{x[ind]}'
    extrafolder = os.listdir(string)

    string = f'{string}/{extrafolder[0]}'


    folderpaths_patients.append(string)
```

```python
dir_all_seg = 'G:/Dokumente/MedicalIM/Lung_Cancer_Seg/manifest-1585232716547/LIDC-IDRI'
all_patients_seg = os.listdir(dir_all_seg)
all_patients_paths_seg = []
ind = 0

for ind in range(len(all_patients_seg)):
    path_ind = os.listdir(dir_all_seg + '/' + all_patients_seg[ind])
    os.listdir(dir_all_seg + '/' + all_patients_seg[ind] + '/' + path_ind[0])

    path_ind = f'{dir_all_seg}/{all_patients_seg[ind]}/{path_ind[0]}'

    x = os.listdir(path_ind)

    #filter all elements from x where segementation is in string
    x = [x[i] for i in range(len(x)) if 'Seg' in x[i]]

    res = [f'{path_ind}/{x[i]}' for i in range(len(x))]

    all_patients_paths_seg.append(res)

all_patients_paths_seg

#G:\Dokumente\MedicalIM\Lung_Cancer_Seg\manifest-1585232716547\LIDC-IDRI\LIDC-IDRI-0001\01-01-2000-NA-NA-30178\3000567.000000-Segmentation of Nodule 1 - Annotation Nodule 001-90548
```
