```python
import pandas as pd
import re
import numpy as np
```

```python
pd.read_excel("tcia-diagnosis-data-2012-04-20.xls")['TCIA Patient ID'].tolist()
```

```python
diag = pd.read_excel("tcia-diagnosis-data-2012-04-20.xls")
diag.replace(r'^\s*$', np.nan, regex=True, inplace=True)
diag.dropna(axis=1, how='all', inplace=True)
diag.fillna(0, inplace=True)
diag
```

```python
diag.columns
```

```python
cols = []
for i in diag.columns:
    #cols.append(re.sub('\n0=.*', '', i))
    cols.append(i.split("\n0")[0].strip().replace("\n"," "))
cols
```

```python
diag.columns = cols
```

```python
diag.to_csv("diagnosis.csv", index=False)
```
